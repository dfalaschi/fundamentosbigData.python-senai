```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Mapeamento do caminho e arquivo
path = 'C:\\Users\\FIC\\Documents\\david_python\\'
file = 'protetor_solar.csv'

# Atribuir o arquivo a um dataframe
protetorSolar = pd.read_csv(path + file)

protetorSolar
```

```python
protetorSolar.columns = ['Temperatura', 'Vendas']
```

```python
sns.jointplot(x = 'Temperatura', y= 'Vendas', data = protetorSolar, color = 'red')
```

```python
sns.pairplot(protetorSolar)
```

```python
sns.lmplot(
    x = 'Temperatura',
    y = 'Vendas',
    data = protetorSolar
           )
```

```python
x = protetorSolar[['Temperatura']]
y = protetorSolar['Vendas']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression(fit_intercept= True)
regressor.fit(X_train, y_train)
```

```python
print('Coeficiente angular (a): ', regressor.coef_)
print('Coficiente linear (b):', regressor.intercept_)
```

```python
y_previsao = regressor.predict(X_test)

y_previsao
```

```python
# Criar um grÃ¡fico de dispersÃ£o

plt.scatter(X_train,y_train, color = 'green')
plt.plot(X_train, regressor.predict(X_train), color = 'red')
plt.ylabel('Receita [Dolares]')
plt.xlabel('Temperatura [Â°C]')
plt.title('Receita gerada Vs. Temperatura (Vendas de Protetor Solar)')
```
