### Exercício 01

Escreva um programa Python para desenhar uma linha com rótulo
adequado no eixo x, eixo y e um título.

```python
# Importar a bibloteca

import matplotlib.pyplot as plt

# Gerar os dados para inserir no gráfico

x = range(1,50)

y = [value * 3 for value in x]
```

```python
print('Valores de X:')
print(range(1,50))
print('Valores de Y (Três vezes o X)')
print(y)
```

```python
plt.plot(x,y)

plt.title('Draw a line', fontsize = 15, c= 'blue')

plt.xlabel("Valores de X")
plt.ylabel('Valores de Y')
```
