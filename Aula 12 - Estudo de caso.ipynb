### Criando um gráfico de colunas

```python
graficos = sns.barplot(x='Categoria',y='ValorVenda', data=df, errorbar=None)

# O método sns.barplot vai retornar uma lista de sub metodos utilizando os containers para acessar o texto dos rótulos para cada coluna, desta forma o python entende que ele deve adicionar um rótulo em cada coluna, de acordo
# com o comando graficos.containers[0]

graficos.bar_label(graficos.containers[0])
```

```python
graficos = sns.barplot(x='Segmento', y='ValorVenda',data=df, errorbar=None)

graficos.bar_label(graficos.containers[0])
```

```python
graficos = sns.barplot(y='Produto', x='ValorVenda', data=df, orient='h', errorbar=None)

graficos.bar_label(graficos.containers[0])
```

```python
graficos = sns.barplot(x='Cidade',y='ValorVenda', data=df, errorbar=None)

graficos.bar_label(graficos.containers[0])
```

```python
graficos = sns.barplot(x='Estado', y='ValorVenda', data=df, errorbar=None)

graficos.bar_label(graficos.containers[0])
```

```python
graficos = sns.barplot(x='Vendedor', y='ValorVenda', data=df, errorbar=None)

graficos.bar_label(graficos.containers[0])
```
