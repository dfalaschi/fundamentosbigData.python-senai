### Importando as Bibliotecas (Matplotlib)

```python
import matplotlib.pyplot as plt   # Permite a geração de graficos
import numpy as np  # Permite trabalhar com matrizes e vetores
```

### Primeiro gráfico com Matplotlib

```python
# Criando os vetores

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Plotar o gráfico

plt.plot(x,y)
```

### Modificando o tipo dos marcadores

```python
plt.plot(x,y, marker = '*')
```

### Alterando as cores das linhas

```python
plt.plot(x,y, marker = '*', c='#A27659')
```

### Mudando como os pontos são ligados e o tipo de linha

```python
plt.plot(
    x,y,  # Os dados do gráfico
    marker = '*',  # Tipo de marcador
    c = 'green', # Cor da linha
    ds = 'steps', # A forma com que os dados são conectados
    ls = '--' # Estilo de linha
)
```

### Matplotlib – adicionando outros elementos

Podemos ainda adicionar labels aos eixos, título ao gráfico, modificar o tamanho da fonte, dentre outras
possibilidades. Tome cuidado para que seu gráfico não tenha muitos elementos e acabe tirando a
atenção do que realmente importa.

```python
# Adiciona dados ao gráfico
plt.plot(x,y)

# Adicionar titulo ao eixo X
plt.xlabel('Valores de X', fontsize = 12)

# Adicionar título ao eixo y
plt.ylabel('Valores de Y', fontsize = 12)

# Adicionar título ao gráfico
plt.title('Título do Gráfico', fontsize = 18, c='purple')

# Alterar as cores dos valores dos eixos
plt.xticks(fontsize = 8, c='blue')

plt.yticks(fontsize = 8, c='green')
```

### Múltiplas curvas no mecanismo gráfico

```python
import matplotlib.pyplot as plt

import numpy as np

# Retorna (Linspace) valores igualmente espaçados em um intervalo

x = np.linspace (0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y)
plt.plot(x,z)
```

### Matplotlib - Múltiplos Subgráficos

```python
# O comando plt.subplot permite criar grupo de gráfico
# plt.subplot (linhas, colunas, Posição elemento)

plt.subplot(2,1,1) # Primeiro gráfico
plt.plot(x,y) # Conjunto de dados do primeiro gráfico
plt.title('subplots') 

plt.subplot(2,1,2) # Segundo gráfico
plt.plot(x,z) # Conjunto de dados do segundo gráfico
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('subplots')
plt.subplot(1,2,2)
plt.plot(x,z)
```

##### Seaborn – Gráficos mais elaborados



Seaborn é uma biblioteca baseada em Matplolib que traz uma interface mais atraente para os gráficos,
além de facilitar a modificação de parâmetros default do Matplotlib e permite a utilização de dataframes.

```python
# O ALIAS(apelido) da biblioteca SEABORN fica como sns

import seaborn as sns
import numpy as np

# Retorna (Linspace) valores igualmente espaçados em um intervalo
x = np.linspace (0,2*np.pi,100) 
y = np.sin(x) 
```

### Seaborn – Datasets embutidos
O Seaborn possui diversos dados embutidos que podem ser utilizados para estudar a própria
biblioteca. Para sabermos quais os datasets estão disponíveis, podemos utilizar a função
get_datasets_names ().

```python
sns.get_dataset_names()
```

Criando um dataframe com o dataset íris.

```python
iris = sns.load_dataset('iris')
iris.head()
```

Criando um dataframe com o dataset planets.

```python
# O dataset planets ou iris já estão integrados no Seaborn
# dataset para estudo ou simulação

planetas = sns.load_dataset('planets')
planetas.head()
```

### Seaborn – Ajustando os gráficos

```python
import matplotlib.pyplot as plt

sns.set()

plt.plot(x,y)
```

```python
# Alterando o tema para claro

sns.set_style('whitegrid')
plt.plot(x,y)
```

```python
# Alterando o tema para preto

sns.set_style('dark')
plt.plot(x,y)
```
