### Exercício 02 

Escreva um programa Python para desenhar uma linha usando
determinados valores de eixo com rótulo adequado no eixo x, eixo y e um título.


```python
import matplotlib.pyplot as plt

x = [1,2,3]

y = [2,4,1]

plt.plot(x,y)

plt.title('Sample graph!', fontsize = 15, c='gray')
plt.xlabel('x - axis', fontsize = 10)
plt.ylabel('y - axis',fontsize=10 )
```

### Exercício 04

 Escreva um programa Python para desenhar gráficos de
linhas dos dados financeiros da Alphabet Inc. entre 3 de outubro de 2016
a 7 de outubro de 2016.
Exemplo de dados estão no arquivo fdata.csv.

```python
# Importar as bibliotecas

import matplotlib.pyplot as plt
import pandas as pd

# Mapear o caminho 

path = 'C:\\Users\\FIC\\Documents\\david_python\\'
file = 'fdata.csv'

# Abrir o arquivo
# O argumento index_col permite slecionar qual colune será o indice do dataframe. Exemplo 0 - Primeira Colune, 1 - Segunda coluna como indice e Nome - padrao
df = pd.read_csv(path + file, sep = ',', index_col = None)

# Visualizar o cabeçalho

df.head()

# Plotar os dados

df.plot()
```

```python
df.head()
df.columns = ['data','abertura','alta','baixa','fechamento']
df.head()
```

```python
df_2 = df[['data','abertura']]

df_2.info()
```

```python
# Substituir uma string
df_2['abertura'] = df_2['abertura'].replace('774. 25', '774.25')

# Converter em numerico pd.to_numeric(coluna)
df_2['abertura'] = pd.to_numeric(df_2['abertura'])
```

```python
x = df_2['data']
y = df_2['abertura']

plt.plot(x,y)
```

```python
df_3 = df[['data','alta']]

df_3.info()

x = df_3['data']
y = df_3['alta']

plt.plot(x,y)
```
